```python
import pandas as pd
from fuzzywuzzy import fuzz

# Đọc tập dữ liệu từ tệp CSV
# df = pd.read_csv('test_data.csv')

# Giả định đã có dữ liệu và các cột 'Evidence_Predict_1' đến 'Evidence_Predict_5'
# y_true là cột chứa dữ liệu evidence thực tế
y_true = df['Evidence']

# Hàm để tính toán độ tương tự
def calculate_similarity(pred, true):
    return fuzz.token_set_ratio(pred, true)

# Hàm để tính độ chính xác cho từng nhãn
def calculate_accuracy_for_label(y_true, y_pred_list, threshold=90):
    labels = ['HỖ TRỢ', 'PHỦ NHẬN']  # Các nhãn cần tính độ chính xác
    label_accuracy = {label: 0 for label in labels}
    total_predictions = len(y_true)

    for i in range(total_predictions):
        true_evidence = y_true[i]
        best_accuracy = 0

        # Tính độ tương tự cho từng dự đoán và nhãn thực tế
        for pred in y_pred_list[i]:
            similarity = calculate_similarity(pred, true_evidence)
            if similarity >= threshold:
                best_accuracy = max(best_accuracy, similarity)

        # Xác định nhãn của evidence thực tế
        true_label = df.loc[i, 'Verdict']

        # Nếu có một trong 5 evidence predict có độ tương tự >= ngưỡng, kiểm tra nhãn và tăng accuracy
        if best_accuracy >= threshold:
            if true_label in labels:
                label_accuracy[true_label] += 1

    # Tính toán độ chính xác cho từng nhãn
    for label in labels:
        label_count = df[df['Verdict'] == label].shape[0]
        if label_count > 0:
            label_accuracy[label] /= label_count

    return label_accuracy

# Tạo một danh sách các cột dự đoán evidence
y_pred_list = df_kq[['evidence_1', 'evidence_2', 'evidence_3', 'evidence_4', 'evidence_5']].values.tolist()

# Tính toán độ chính xác cho từng nhãn
accuracy_per_label = calculate_accuracy_for_label(y_true, y_pred_list, threshold=100)

# In kết quả độ chính xác cho từng nhãn
for label, acc in accuracy_per_label.items():
    print(f'Độ chính xác nhãn {label}: {acc * 100:.2f}%')
```
