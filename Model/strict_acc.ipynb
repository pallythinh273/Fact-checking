# PhoBERT + BM25

```python
vp = pd.read_csv('../Data/Result_VP_PhoBert.csv')
vp.head()
```

```python
len(vp)
```

```python
er = pd.read_csv('../Data/BM25_output_with_top5_evidence.csv')
er.head()
```

```python
len(er)
```

```python
match_df = pd.merge(vp, er, on='Claim_ID', how='inner')
len(match_df)
```

```python
match_df.columns
```

```python
match_df.drop(columns=['Claim_x', 'Context_ID_x', 'Context_x', 'Claim_y', 'Context_ID_y', 'Context_y', 'Verdict_y', 'Evidence_y'], inplace=True)
match_df.head()
```

```python
match_df.shape
```

```python
match_df['Verdict_label'] = match_df['Verdict_x'].apply(lambda x: 0 if x == 'HỖ TRỢ' else 1)
match_df.drop(columns=['Verdict_x'], inplace=True)
match_df.head()
```

```python
count = 0
from fuzzywuzzy import fuzz

# Hàm để tính toán độ tương tự
def calculate_similarity(pred, true):
    return fuzz.token_set_ratio(pred, true)


for i in range(match_df.shape[0]):
    # check verdict prediction
    verdict = 0
    if match_df['predict_label'][i] == match_df['Verdict_label'][i]:
        verdict = 1

    # check evidence retrieval
    evidence = 0
    y_true = match_df['Evidence_x'][i]
    y_pred_list = [match_df['Evidence_Predict_1'][i], match_df['Evidence_Predict_2'][i], match_df['Evidence_Predict_3'][i], match_df['Evidence_Predict_4'][i], match_df['Evidence_Predict_5'][i]]
    for pred in y_pred_list:
        if calculate_similarity(pred, y_true) == 100:
            evidence = 1
            break
    
    if verdict == 1 and evidence == 1:
        count += 1

strict_accuracy = count / match_df.shape[0]
print(f"Strict accuracy: {strict_accuracy*100:.2f} %")
```
